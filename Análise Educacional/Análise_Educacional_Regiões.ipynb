```python
# Filtrar os dados para 'todos (1-4)'
media_finais = df[df['anos_escolares'] == 'todos (1-4)']

# Filtrar para a regiÃ£o 'Sudeste'
media_finais_nordeste = media_finais[media_finais['regiao'] == 'Sudeste']

# Calcular a mÃ©dia da 'taxa_aprovacao' por 'rede'
mediaTA_por_rede = media_finais_nordeste.groupby('rede')['taxa_aprovacao'].mean()

# Ordenar os valores
mediaTA_por_rede = mediaTA_por_rede.sort_values()

# Criar um grÃ¡fico de barras
mediaTA_por_rede.plot(kind = 'bar', figsize = (10, 6), color = 'yellow')

plt.xlabel('Rede de Ensino')
plt.ylabel('MÃ©dia da Taxa de AprovaÃ§Ã£o')
plt.title('Taxa de AprovaÃ§Ã£o por Rede de Ensino na RegiÃ£o Sudeste para o Ensino MÃ©dio')
plt.ylim(70, 100)
plt.savefig('EM-Sudeste.png')
plt.show()
```
